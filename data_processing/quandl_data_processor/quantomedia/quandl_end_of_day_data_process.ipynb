# Write

```python
import pandas as pd
```

```python
frame = pd.read_csv(r"D:\data\quandl\stock\day\EOD_20200627.csv", header=None)
```

```python
columns = ["symbol", "date", "unadj_open", "unadj_high", "unadj_low", "unadj_close", "unadj_vol", "dividends", "splits", "adj_open", "adj_high", "adj_low", "adj_close", "adj_volume"]
```

```python
frame.columns = columns
```

```python
frame.loc[:, "date"] = pd.to_datetime(frame['date'], format='%Y-%m-%d')
```

```python
frame.set_index("date", drop=True, inplace=True)
```
