```python
import paratext
import pandas as pd
```

```python
def fast_read_csv(path):
    df = paratext.load_csv_to_pandas(path, cat_names = ['ticker', 'date', 'lastupdated'])

    df.set_index(0, inplace=True, drop=True)

    df_correct_dict = {}
    for column_name in df.index.tolist():
        df_correct_dict[column_name] = df.loc[column_name, :].tolist()[0]

    df_correct = pd.DataFrame(df_correct_dict)

    for index, type_value in df_correct.dtypes.iteritems():
        if type_value == object:
            df_correct.loc[:, index] = df_correct.loc[:, index].apply(lambda x: x.decode("utf-8") )

    return df_correct
```
