```python
import ondemand

od = ondemand.OnDemandClient(api_key='')
```

```python
import datetime
import time
import numpy as np
import pandas as pd
import logging
%matplotlib inline
```

```python
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(name)s - %(levelname)s - %(message)s')
logger = logging.getLogger("barchart")
handler = logging.FileHandler('barchart_downloader.log')
formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
handler.setFormatter(formatter)
logger.addHandler(handler)
```

```python
symbol_df = pd.read_csv(r"D:\my_interested_symbols.csv")
```

```python
year_list = range(30)
```

```python
symbols = [];
for ix, row in symbol_df.iterrows():
    for month_code in row.loc["Months"]:
        for year_code in year_list:
            new_symbol = row["Symbol"] + month_code + str(year_code)
            symbols.append(new_symbol)
```

```python
len(symbols)
```

```python
number = 0
for symbol in symbols:
    number += 1
    download_successful = False
    while not download_successful:
        try:
            resp = od.history(symbol, 'minutes', startDate='20020601')
            time.sleep(4)
            if resp["results"]:
                logger.info("Successfully download: {}".format(symbol))
                logger.info("Successfully download: {}/{}".format(number,len(symbols)))
                frame = pd.DataFrame(resp['results'])
                frame.to_csv(symbol + ".csv", index=False)
            else:
                logger.info("Invalid symbol: {}".format(symbol))
            download_successful = True
        except:
            time.sleep(30)
```
